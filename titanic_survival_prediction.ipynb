```python
import numpy as np;
import pandas as pd;

from sklearn.model_selection import train_test_split;
from sklearn.impute import SimpleImputer;
from sklearn.preprocessing import OneHotEncoder;
from sklearn.preprocessing import MinMaxScaler;
from sklearn.tree import DecisionTreeClassifier;


```

```python
df= pd.read_csv('/content/train.csv')
```

```python
df.drop(columns=['PassengerId','Name','Cabin','Ticket'],inplace=True)
```

```python
df.head()
```

```python
x_train, x_test, y_train,y_test= train_test_split(df.drop(columns=['Survived']), df['Survived'],test_size=0.2,random_state=8);
```

```python
y_train.head(2);
```

```python
si_Age= SimpleImputer()
si_embarked= SimpleImputer(strategy= 'most_frequent')

x_train_Age= si_Age.fit_transform(x_train[['Age']]);
x_train_embarked= si_embarked.fit_transform(x_train[['Embarked']]);

x_test_Age= si_Age.fit_transform(x_test[['Age']]);
x_test_embarked= si_embarked.fit_transform(x_test[['Embarked']]);
```

```python
ohe_sex= OneHotEncoder(sparse=False,handle_unknown='ignore');
ohe_embarked= OneHotEncoder(sparse=False, handle_unknown='ignore')

x_train_sex=ohe_sex.fit_transform(x_train[['Sex']]);
x_train_embarked= ohe_embarked.fit_transform(x_train_embarked);

x_test_sex= ohe_sex.fit_transform(x_test[['Sex']]);
x_test_embarked= ohe_embarked.fit_transform(x_test_embarked);
```

```python
x_train_rem=x_train.drop(columns=['Sex','Embarked','Age'])
x_test_rem= x_test.drop(columns=['Sex','Embarked','Age'])
```

```python
x_train_transformed= np.concatenate((x_train_Age,x_train_embarked,x_train_sex,x_train_rem),axis=1);
x_test_transformed= np.concatenate((x_test_Age,x_test_embarked,x_test_sex,x_test_rem),axis=1);
```

```python
clf= DecisionTreeClassifier();

clf.fit(x_train_transformed,y_train);
```

```python
y_pred= clf.predict(x_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
